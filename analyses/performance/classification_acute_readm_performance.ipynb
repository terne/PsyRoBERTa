```python
# Results plot for paper

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
sns.set_style("whitegrid")

labels = ["PsyRoBERTa",
          "MeDa-BERT",
          "RøBÆRTa",
          "BERT",  
          "LogReg"] # LogReg results were computed in ../logistic_regression/*.ipynb, while the were computed above.

# models1 = all notes
models1_auroc = [[0.736, 0.727, 0.745],
                [0.734, 0.725, 0.743],
                [0.728, 0.719, 0.737],
                [0.718, 0.708, 0.727],
                [0.718, 0.709,0.727]
                ]

models1_mcc = [[0.303, 0.287,0.322],
               [0.295, 0.278,0.312],
               [0.263, 0.245,0.282],
               [0.193, 0.175,0.212],
               [0.258, 0.245,0.271]]

models1_auprc = [[0.464, 0.446,0.482],
                 [0.458, 0.44,0.474],
                 [0.447, 0.43,0.464],
                 [0.420, 0.402,0.436],
                 [0.422, 0.405,0.438]]

models1_F1_avg = [[0.770, 0.763,0.777],
                  [0.767, 0.760,0.774],
                  [0.756, 0.748,0.763],
                  [0.729, 0.721,0.738],
                  [0.622, 0.616,0.629]]

# models2 = discharge summaries
models2_auroc = [[0.711, 0.702, 0.721],
         [0.700, 0.691, 0.711],
         [0.689, 0.679, 0.698],
         [0.679, 0.669, 0.689],
         [0.680, 0.670, 0.691]
         ]

models2_mcc = [[0.285, 0.269,0.304],
               [0.264, 0.248,0.283],
               [0.226, 0.208,0.244],
               [0.215, 0.198,0.231],
               [0.223, 0.208,0.239]
               ]

models2_auprc = [[0.420, 0.403,0.437],
                 [0.399, 0.385,0.416],
                 [0.378, 0.362,0.394],
                 [0.368, 0.353,0.383],
                 [0.390, 0.374,0.407]
                 ]

models2_F1_avg = [[0.759, 0.752,0.766],
                  [0.745, 0.739,0.753],
                  [0.747, 0.74,0.755],
                  [0.742, 0.735,0.75],
                  [0.691, 0.684,0.697]
                  ]



def test_metric_with_intervals(models1,models2,labels, ylabel):

    colors = sns.color_palette("mako_r",n_colors=len(models1))

    fig, ax = plt.subplots()

    ax.set_xticks(np.arange(0,len(models1),1));
    ax.set_xticklabels(labels)
    ax.set_ylabel(ylabel)


    for indel, (metric_all,metric_discharge) in enumerate(list(zip(models1,models2))):
        
    
        res1 = metric_all[0]
        res2 = metric_discharge[0]


       
        low_conf1, high_conf1 = metric_all[1], metric_all[2]
        
    
        low_conf2, high_conf2 = metric_discharge[1], metric_discharge[2]
 

        horizontal_line_width = 0.15
        adjust = 0.1

        left1 = indel-adjust - horizontal_line_width / 2
        right1 = indel-adjust + horizontal_line_width / 2
        
        left2 = indel+adjust - horizontal_line_width / 2
        right2 = indel+adjust + horizontal_line_width / 2

        ax.plot([indel-adjust, indel-adjust], [low_conf1, high_conf1],color=colors[0])
        ax.plot([left1, right1], [high_conf1, high_conf1],color=colors[0])
        ax.plot([left1, right1], [low_conf1, low_conf1], color=colors[0])
        ax.plot(indel-adjust,res1, "o",color=colors[0])
        
        ax.plot([indel+adjust, indel+adjust], [low_conf2, high_conf2],color=colors[2])
        ax.plot([left2, right2], [high_conf2, high_conf2],color=colors[2])
        ax.plot([left2, right2], [low_conf2, low_conf2], color=colors[2])
        ax.plot(indel+adjust,res2, "o",color=colors[2])
        
        custom_lines = [
            Line2D([1], [1], color=colors[0]),
            Line2D([1], [1], color=colors[2])]

        
        ax.legend(custom_lines, ['All admission notes', 'Discharge summaries'], bbox_to_anchor=(1.5, 1))
```

```python
test_metric_with_intervals(models1_auroc,models2_auroc,labels, "AUROC")
plt.savefig("../../output/testperf_bootstrap_AUROC.pdf", bbox_inches='tight')
```

```python
test_metric_with_intervals(models1_auprc,models2_auprc,labels, "AUPRC")
plt.savefig("../../output/testperf_bootstrap_AUPRC.pdf", bbox_inches='tight')
```

```python
test_metric_with_intervals(models1_F1_avg,models2_F1_avg,labels, "Weighted average F1")
plt.savefig("../../output/testperf_bootstrap_F1AVG.pdf", bbox_inches='tight')
```
